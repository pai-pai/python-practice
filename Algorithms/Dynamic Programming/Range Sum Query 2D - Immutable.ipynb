Given a 2D matrix $\textit{matrix}$, handle multiple queries of the following type:

- Calculate the **sum** of the elements of matrix inside the rectangle defined by its **upper left corner** (row1, col1) and **lower right corner** (row2, col2).

Implement the NumMatrix class:

- `NumMatrix(int[][] matrix)` Initializes the object with the integer matrix matrix.
- `int sumRegion(int row1, int col1, int row2, int col2)` Returns the sum of the elements of matrix inside the rectangle defined by its upper left corner (row1, col1) and lower right corner (row2, col2).


*Example 1*


$\textit{matrix}$ = [[3, 0, 1, 4, 2], [5, 6, 3, 2, 1], [1, 2, 0, 1, 5], [4, 1, 0, 1, 7], [1, 0, 3, 0, 5]]<br />
- row1 = 2, col1 = 1, row2 = 4, col2 = 3
- row1 = 1, col1 = 1, row2 = 2, col2 = 2
- row1 = 1, col1 = 2, row2 = 2, col2 = 4

Returns:
- 8 (i.e sum of the red rectangle)
- 11 (i.e sum of the green rectangle)
- 12 (i.e sum of the blue rectangle)


*Code*

```python
class NumMatrix:

    def __init__(self, matrix):
        rows, cols = len(matrix), len(matrix[0])
        self.summed_matrix = [[0] * (cols+1) for _ in range(rows+1)]
        for r in range(rows):
            for c in range(cols):
                self.summed_matrix[r+1][c+1] = (self.summed_matrix[r+1][c] + self.summed_matrix[r][c+1] -
                                                self.summed_matrix[r][c] + matrix[r][c])

    def sumRegion(self, row1, col1, row2, col2):
        return (self.summed_matrix[row2+1][col2+1] - self.summed_matrix[row2+1][col1] -
                self.summed_matrix[row1][col2+1] + self.summed_matrix[row1][col1])
```

```python
matrix = NumMatrix([[3, 0, 1, 4, 2], [5, 6, 3, 2, 1], [1, 2, 0, 1, 5], [4, 1, 0, 1, 7], [1, 0, 3, 0, 5]])
```

```python
matrix.sumRegion(2, 1, 4, 3)
```

```python
matrix.sumRegion(1, 1, 2, 2)
```

```python
matrix.sumRegion(1, 2, 2, 4)
```
