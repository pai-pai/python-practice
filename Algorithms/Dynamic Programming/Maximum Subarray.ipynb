Given an integer array $\textit{nums}$, find the contiguous subarray (containing at least one number) which has the largest sum and return its sum.

A **subarray** is a **contiguous** part of an array.


*Example 1*

$\textit{nums}$ = [-2,1,-3,4,-1,2,1,-5,4]<br />
Returns 6. [4,-1,2,1] has the largest sum = 6.

*Example 2*

$\textit{nums}$ = [1]<br />
Returns 1

*Example 3*

$\textit{nums}$ = [5,4,-1,7,8]<br />
Returns 23


*Code*

```python
class Solution:
    def maxSubArray(self, nums):
        total = nums[0]
        current_sum = 0
        for num in nums:
            current_sum += num
            total = max(total, current_sum)
            if current_sum < 0:
                current_sum = 0
        return total
```

```python
print(Solution().maxSubArray([-2,1,-3,4,-1,2,1,-5,4]))
```

```python
print(Solution().maxSubArray([1]))
```

```python
print(Solution().maxSubArray([5,4,-1,7,8]))
```

```python
print(Solution().maxSubArray([-1]))
```
