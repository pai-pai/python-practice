Given the $head$ of a singly linked list, return the middle node of the linked list.

If there are two middle nodes, return **the second middle** node.


*Example 1*

$head$ = [1,2,3,4,5]<br />
Returns [3,4,5]. The middle node of the list is node $3$.

*Example 2*

$head$ = [1,2,3,4,5,6]<br />
Returns [4,5,6]. Since the list has two middle nodes with values 3 and 4, we return the second one.


*Code*

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def list_to_linked_list(l):
    if not l:
        return
    result = current = ListNode(l[0])
    for el in l[1:]:
        current.next = ListNode()
        current = current.next
        current.val = el
    return result

def middle_node(head):
    slow = fast = head
    while fast and fast.next:
        slow = slow.next
        fast = fast.next.next
    return slow
```

```python
head = list_to_linked_list([1,2,3,4,5])
result = middle_node(head)
output = []
while result:
    output.append(result.val)
    result = result.next
print(output)
```

```python
head = list_to_linked_list([1,2,3,4,5,6])
result = middle_node(head)
output = []
while result:
    output.append(result.val)
    result = result.next
print(output)
```
