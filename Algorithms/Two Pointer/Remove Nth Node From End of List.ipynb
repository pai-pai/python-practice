Given the $head$ of a linked list, remove the $n^{th}$ node from the end of the list and return its head.


*Example 1*


$head$ = [1,2,3,4,5], $n$ = 2<br />
Returns [1,2,3,5]

*Example 2*

$head$ = [1], $n$ = 1<br />
Returns []

*Example 3*

$head$ = [1,2], $n$ = 1<br />
Returns [1]

*Code*

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def list_to_linked_list(l):
    if not l:
        return
    result = current = ListNode(l[0])
    for el in l[1:]:
        current.next = ListNode()
        current = current.next
        current.val = el
    return result

def remove_nth_from_end(head, n):
    slow = fast = head
    # The fast pointer starts from nth node because in this case 
    # the slow pointer stops exactly at nth node
    for _ in range(n):
        fast = fast.next
    if not fast:
        return head.next
    while fast.next:
        fast, slow = fast.next, slow.next
    slow.next = slow.next.next
    return head
```

```python
head = list_to_linked_list([1,2,3,4,5])
result = remove_nth_from_end(head, 2)
output = []
while result:
    output.append(result.val)
    result = result.next
print(output)
```

```python
head = list_to_linked_list([1])
result = remove_nth_from_end(head, 1)
output = []
while result:
    output.append(result.val)
    result = result.next
print(output)
```

```python
head = list_to_linked_list([1,2])
result = remove_nth_from_end(head, 1)
output = []
while result:
    output.append(result.val)
    result = result.next
print(output)
```
