### Permutations and Combinations

Permutatuon ($P$) is a rearrangement of elements of an ordered set. For combination ($C$) the he order does not matter.

$$_{n}P_{r} = \frac{n!}{(n - r)!}$$
$$_{n}C_{r} = \frac{_{n}P_{r}}{r!} = \frac{n!}{(n - r)!\cdot r!}$$

___
*Example 1*

Given set $\{A, B, C, D\}$.

Write down all permutations of 2 elements for this set (underlined elements are duplicated **combinations**):
$$
 \begin{matrix}
  AB & \underline{BA} & \underline{CA} & \underline{DA} \\
  AC & BC & \underline{CB} & \underline{DB} \\
  AD & BD & CD & \underline{DC}
 \end{matrix}
$$

Calculate permutations and combinations:
$$_{4}P_{2} = \frac{4!}{(4 - 2)!} = \frac{4!}{2!} = 12$$
$$_{4}C_{2} = \frac{4!}{(4 - 2)!\cdot 2!} = \frac{4!}{2!\cdot 2!} = 6$$

___
*Example 2*

In how many different ways can you arrange $3$ books on a shelf from a group of $7$?

It is a permutation because order does matter here:
$$_{7}P_{3} = \frac{7!}{(7 - 3)!} = \frac{7!}{4!} = 210$$

___
*Example 3*

In how many different ways can you arrange $5$ books on a shelf?

It is a permutation again:
$$_{5}P_{5} = \frac{5!}{(5 - 5)!} = \frac{5!}{0!} = 120$$
Here $0! = 1$!

___
*Example 4*

How many teams of 4 can be produced from a pool of 12 engineers?

It is a combination because order does not matter (because it does not matter order of engineers in a team):
$$_{12}C_{4} = \frac{12!}{(12 - 4)!\cdot 4!} = \frac{12!}{8!\cdot 4!} = 495$$

___
*Example 5*

In how many different ways can you arrange the letters in the word 'ALABAMA'?

- Count a number of a letters in the word: $7$.
- Letter 'A' repeats $4$ times.
$$\frac{7!}{4!} = 210$$

___
*Example 6*

In how many different ways can you arrange the letters in the word 'MISSISSIPPI'?
- Count a number of a letters in the word: $11$.
- Letter 'I' repeats $4$ times.
- Letter 'S' repeats $4$ times.
- Letter 'P' repeats $2$ times.
$$\frac{11!}{4!\cdot 4!\cdot 2!} = 34650$$
