You are given two **non-empty** linked lists representing two non-negative integers. The digits are stored in **reverse order**, and each of their nodes contains a single digit. Add the two numbers and return the sum as a linked list.

You may assume the two numbers do not contain any leading zero, except the number 0 itself.

*Example 1*

<br />
$l1 = [2,4,3] \\
l2 = [5,6,4]$
Returns $[7,0,8]$. Because $342 + 465 = 807$.

*Example 2*

$l1 = [0] \\
l2 = [0]$
Returns $[0]$.

*Example 3*

$l1 = [9,9,9,9,9,9,9] \\
l2 = [9,9,9,9]$
Returns $[8,9,9,9,0,0,0,1]$.

*Code*

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def list_to_linked_list(l):
    if not l:
        return
    result = current = ListNode(l[0])
    for el in l[1:]:
        current.next = ListNode()
        current = current.next
        current.val = el
    return result

def add_two_numbers(l1, l2):
    result = current = ListNode()
    while l1 or l2:
        if l1:
            current.val += l1.val
            l1 = l1.next
        if l2:
            current.val += l2.val
            l2 = l2.next
        remains = current.val // 10
        current.val %= 10
        if l1 or l2 or remains:
            current.next = ListNode(remains)
            current = current.next
    return result
```

```python
l1 = list_to_linked_list([2,4,3])
l2 = list_to_linked_list([5,6,4])
result = add_two_numbers(l1, l2)
output = []
while result:
    output.append(result.val)
    result = result.next
print(output)
```

```python
l1 = list_to_linked_list([0])
l2 = list_to_linked_list([0])
result = add_two_numbers(l1, l2)
output = []
while result:
    output.append(result.val)
    result = result.next
print(output)
```

```python
l1 = list_to_linked_list([9,9,9,9,9,9,9])
l2 = list_to_linked_list([9,9,9,9])
result = add_two_numbers(l1, l2)
output = []
while result:
    output.append(result.val)
    result = result.next
print(output)
```
