Given the $\textit{head}$ of a singly linked list, return `true` if it is a palindrome.


*Example 1*


$\textit{head}$ = [1,2,2,1]<br />
Returns `true`

*Example 2*


$\textit{head}$ = [1,2]<br />
Returns `false`


*Code*

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def list_to_linked_list(l):
    if not l:
        return
    result = current = ListNode(l[0])
    for el in l[1:]:
        current.next = ListNode()
        current = current.next
        current.val = el
    return result

def is_palindrome(head):
    slow, fast = head, head
    while fast and fast.next:  # fast.next check is needed for odd length case
        slow = slow.next
        fast = fast.next.next
    stack = []
    while slow:
        stack.append(slow.val)
        slow = slow.next
    while stack:
        if head.val != stack.pop():
            return False
        head = head.next
    return True
```

```python
l = list_to_linked_list([1,2,3,2,1])  # exp True
print(is_palindrome(l))
```

```python
l = list_to_linked_list([1,2,2,1])  # exp True
print(is_palindrome(l))
```

```python
l = list_to_linked_list([1])  # exp True
print(is_palindrome(l))
```

```python
l = list_to_linked_list([1,2,3])  # exp False
print(is_palindrome(l))
```
