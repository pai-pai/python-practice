You are given the heads of two sorted linked lists $list1$ and $list2$.

Merge the two lists in a one **sorted** list. The list should be made by splicing together the nodes of the first two lists.

Return **the head of the merged linked list**.


*Example 1*


$list1$ = [1,2,4]<br />
$list2$ = [1,3,4]<br />
Returns [1,1,2,3,4,4]

*Example 2*

$list1$ = []<br />
$list2$ = []<br />
Returns []

*Example 3*

$list1$ = []<br />
$list2$ = [0]<br />
Returns [0]


*Code*

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def list_to_linked_list(l):
    if not l:
        return
    result = current = ListNode(l[0])
    for el in l[1:]:
        current.next = ListNode()
        current = current.next
        current.val = el
    return result

def merge_two_lists(list1, list2):
    if not list1 or not list2:
        return list1 or list2
    recipient, donor = (list1, list2) if list1.val <= list2.val else (list2, list1)
    merged = recipient
    while donor:
        if not recipient.next:
            recipient.next = donor
            return merged
        if donor.val < recipient.next.val:
            rtemp = recipient.next
            dtemp = donor.next
            recipient.next = donor
            recipient.next.next = rtemp
            donor = dtemp
        recipient = recipient.next
    return merged
```

```python
l1 = list_to_linked_list([1,2,4])
l2 = list_to_linked_list([1,3,4])
result = merge_two_lists(l1, l2)
output = []
while result:
    output.append(result.val)
    result = result.next
print(output)
```

```python
l1 = list_to_linked_list([])
l2 = list_to_linked_list([])
result = merge_two_lists(l1, l2)
output = []
while result:
    output.append(result.val)
    result = result.next
print(output)
```
